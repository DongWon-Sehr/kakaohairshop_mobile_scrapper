```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
pd.set_option('display.max_columns',50) # DataFrame truncation 없이 보기
pd.set_option('display.max_rows',100)
import glob
import re
from ast import literal_eval
```

```python
paths = glob.glob('./data/shop_existing_check/search/key*')
len(paths)
```

```python
files = []
for path in paths:
    with open(path, 'rb') as f:
        files.append(f.read().decode('utf-8'))
print('file count :', len(files))
```

```python
def tf_capitalize(files: list):
    return [file.replace('true', 'True').replace('false', 'False') for file in files]
```

```python
def remove_emoji(files: list):
    return [re.sub('\\\\[a-zA-Z0-9]+', ' ', file) for file in files]
```

```python
def string2dict(files: list):
    return [literal_eval(file) for file in files]
```

```python
# true, false => True, False로 변환
files = tf_capitalize(files)
```

```python
# 이모지 제거
files = remove_emoji(files)
```

```python
# string 객체를 dict 객체로 변환
files = string2dict(files)
```

```python
search_names = list(search_data.name.values)
input_names = list(input_data.store_name.values)

searchval_inputindexes = []
for search_index, search_name in enumerate(search_names):
    tmp_lst=[]
    for input_index, input_name in enumerate(input_names):
        if search_name in input_name:
            tmp_lst.append(input_index)
    searchval_inputindexes.append({search_data.iloc[search_index].fullName : tmp_lst})
        
print('길이 :', len(searchval_inputindexes))
```

```python
searchval_inputindexes
```

```python
def get_matched_input_index(search_index: int, data: dict):
    result={}
    if list(data.values())[0]:
        for input_index in list(data.values())[0]:
            tmp_lst=[]
            if search_data.fullName[search_index] == input_data.store_name[input_index]:
                tmp_lst.append(input_index)
                result={list(data.keys())[0] : tmp_lst}
            elif search_data.branchName[search_index]:
                if search_data.branchName[search_index] in input_data.store_name[input_index]:
                    tmp_lst.append(input_index)
                    result={list(data.keys())[0] : tmp_lst}
            else:
                pass
    
    if not result:
        result={list(data.keys())[0] : []}
    
    return result
```

```python
matched_searchval_inputindexes=[]
for search_index, data in enumerate(searchval_inputindexes):
    matched_searchval_inputindexes.append(get_matched_input_index(search_index, data))

len(matched_searchval_inputindexes)
```

```python
# 아직 인덱스가 2개 이상인 데이터가 있는지 확인
count=0
for data in matched_searchval_inputindexes:
    if list(data.values())[0]:
        pass
    elif len(list(data.values())[0]) > 1:
        count+=1
        
count
```

```python
search_data['input_index'] = np.nan
for search_index, data in enumerate(matched_searchval_inputindexes):
    if list(data.values())[0]:
        search_data['input_index'][search_index] = list(data.values())[0][0]
```
